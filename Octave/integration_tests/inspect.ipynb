### 0.2. Project root setup

```python
def find_project_root(start: Path | None = None) -> Path:
    """
    Find the project root by walking upward from the current directory.

    The project root is identified by the presence of pyproject.toml.
    """
    current = Path.cwd() if start is None else start.resolve()

    for path in [current, *current.parents]:
        if (path / "pyproject.toml").exists():
            return path

    raise FileNotFoundError("Could not find project root containing pyproject.toml.")
```

```python
PROJECT_ROOT = find_project_root()
```
